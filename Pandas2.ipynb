ðŸ“˜ Pandas GroupBy 
The groupby() function in Pandas is one of the most powerful tools for data analysis. It allows you to:

Split data into groups

Apply functions to each group

Combine the processed results

This approach enables efficient aggregation, transformation, and filtering of large datasets.

ðŸ”¹ The 3-Step GroupBy Process
1ï¸âƒ£ Split â†’ 2ï¸âƒ£ Apply â†’ 3ï¸âƒ£ Combine
STEP 1: Splitting Data into Groups

You can group data by one or more columns (keys).

1. Grouping by a Single Column
df.groupby('Name')
print(df.groupby('Name').groups)


This returns a dictionary where:

keys = unique values in the Name column

values = row indices belonging to each group

Example Output:
{'Abhi': [7], 'Anuj': [1, 5], 'Gaurav': [4], 'Jai': [0, 2], 'Princi': [3, 6]}


To see the first record from each group:

gk = df.groupby('Name')
gk.first()

2. Grouping by Multiple Columns
df.groupby(['Name', 'Qualification']).groups

Output Example:
('Jai', 'MCA'): [2], ('Jai', 'Msc'): [0], ...


Each key is now a tuple representing a multi-column category.

3. Grouping with Sorted or Unsorted Keys

By default, keys are sorted:

df.groupby('Name')['Age'].sum()


To keep the original order:

df.groupby('Name', sort=False)['Age'].sum()

4. Grouping Using Object Attributes

Group values and their indices:

df.groupby('Name').groups

5. Iterating Through Groups
grp = df.groupby('Name')
for name, group in grp:
    print(name)
    print(group)
    print()


Works for multiple keys as well.

6. Selecting a Single Group
grp = df.groupby('Name')
grp.get_group('Jai')


For multi-column groups:

grp = df.groupby(['Name', 'Qualification'])
grp.get_group(('Jai', 'Msc'))

STEP 2: Applying Functions to Groups

You can apply:

Aggregation

Transformation

Filtration

1. Aggregation
Simple Aggregation
grp1 = df.groupby('Name')
grp1['Age'].aggregate(np.sum)

Multiple Aggregations
grp['Age'].agg([np.sum, np.mean, np.std])

Aggregation with Multiple Group Keys
df.groupby(['Name', 'Qualification'])['Age'].aggregate(np.sum)

2. Transformation

This performs group-wise calculations but returns output with the same shape as the original dataset.

Example: Standardizing values inside each group

sc = lambda x: (x - x.mean()) / x.std()
grp2 = df2.groupby('Name')
grp2['Age'].transform(sc)


Useful for:

Normalization within groups

Filling missing values group-wise

Creating new transformed features

3. Filtration

Filter out groups that donâ€™t meet a condition.

Example: Keep groups with at least 2 records:

grp2 = df2.groupby('Name')
grp2.filter(lambda x: len(x) >= 2)

STEP 3: Combining Results

After applying functions, results are returned as DataFrames or Series.

Example: Summarizing by group

df.groupby('Name').agg({'Age': 'sum'})

```python
#Pandas groupby

import pandas as pd

data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
df = pd.DataFrame(data1)
print(df)


df.groupby('Name')
print(df.groupby('Name').groups)
```

```python
#now we can print fiest entries in all grp formed
gk = df.groupby('Name')
gk.first()
```

```python
# Grouping data wt multiplr keys
df.groupby(['Name', 'Qualification'])
print(df.groupby(['Name', 'Qualification']).groups)
```

```python
# grping data by sorting keys
df.groupby('Name')['Age'].sum()
```

```python
#Now we apply groupby() without using sort(we pass sort=False).
df.groupby('Name', sort=False)['Age'].sum()
```

```python
#Grouping data with object attributes
df.groupby('Name').groups
```

```python
# The dict is the result of df.groupby('Name').groups
# - Keys: unique values from the "Name" column ('Abhi', 'Anuj', ...)
# - Values: lists of DataFrame index positions (row labels) that belong to each group
# Example meanings:
#   'Anuj': [1, 5]  -> rows with index 1 and 5 in df have Name == 'Anuj'
#   'Abhi': [7]     -> row index 7 in df has Name == 'Abhi'
#
# To view the actual rows for a group (df and gk are defined in the notebook):
df.loc[gk.groups['Anuj']]
# or
gk.get_group('Anuj')
```

```python
# Iterating through groups
grp = df.groupby('Name')
for name, group in grp:
    print(name)
    print(group)    
    print()
```

```python
#Now we iterate an element of group containing multiple keys.
grp = df.groupby(['Name', 'Qualification'])
for name, group in grp:
    print(name)
    print(group)
    print()
```

 Selecting a group: In order to select a group, we can select group using GroupBy.get_group(). We can select a group by applying a function GroupBy.get_group this function select a single group. 

```python
grp = df.groupby('Name')
grp.get_group('Jai')
```

```python
#we select an object grouped on multiple columns.
grp = df.groupby(['Name', 'Qualification'])
grp.get_group(('Anuj', 'MA'))
```

Applying Functions to Groups
 Aggregation: It allows us to find a summary statistic for each group like summing or averaging values.






```python
grp1 = df.groupby('Name')
grp1['Age'].aggregate(np.sum)
```

2. Transformation: It is a process in which we perform some group-specific computations and return a result with the same index as the original data.

```python
# create df2 from the existing data2 dict and run the group-wise transform
df2 = pd.DataFrame(data2)

grp2 = df2.groupby('Name')
sc = lambda x: (x - x.mean()) / x.std()
df2['Age_std_by_name'] = grp2['Age'].transform(sc)

print(df2[['Name', 'Age', 'Age_std_by_name']])
```

3. Filtration: It is a process which is used to discard groups based on a condition. For example we can filter out groups where the number of records is less than a certain threshold.

```python
grp2 = df2.groupby('Name')
grp2.filter(lambda x: len(x) >= 2)
```

Combining Results :- 

After applying the functions, results are combined back into a data structure like a DataFrame or Series. We can further analyze or visualize the data in its new form.

```python
df.groupby('Name').agg({'Age': 'sum'})
```
