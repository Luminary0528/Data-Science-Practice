```python
import pandas as pd
```

```python
#reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

```python
#adding a new column 
#1. assing()
#   when to use assing ?
#* add multiple columns in a single line of code 
#* When you need to overwrite the values of an existing columns (best practice)

#it returns a new object (a copy) with all the original columns in addition to the new ones'''
```

```python
import numpy as np
```

```python
#create random numbers for our new score columns 
score1 = np.random.randint(1, 100, size=1000)
score2 = np.random.randint(1, 100, size=1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score2, index=np.arange(0, 1000))
```

```python

series1
```

```python
#using assign() to add multiple columns
df_exams.assign(score1=series1, score2=series2)
```

```python
#using assign() to add multiple columns
df_exams = df_exams.assign(score1=series1, score2=series2) #for updating over here dataframe
```

```python
# show dataframe
df_exams
```

```python
#insert
#using index() to add column at a specific position
df_exams.insert(1, 'test', series1)
```

```python
df_exams
```

```python
#Sort a dataframe
```

```python
# sort by one column
df_exams.sort_values('math score')
```

```python
#sort descending by one column
df_exams.sort_values('math score', ascending=False)
```

```python
#sort descending by multiple columns
df_exams.sort_values(['math score', 'reading score'], ascending=False)
```

```python
#sort descending by multiple columns and update dataframe  , for the just copy we use inplace bt for the update go wt the =True 
df_exams.sort_values(['math score', 'reading score'], ascending=False,
                    inplace=True)
```

```python
# sort descending with a key fun
df_exams.sort_values('race/ethnicity', ascending=True,
                    key=lambda col:col.str.lower())
```
